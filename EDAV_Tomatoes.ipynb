```python
import csv
from rotten_tomatoes_scraper.rt_scraper import MovieScraper
from rotten_tomatoes_scraper.rt_scraper import DirectorScraper

movie_list = open('Movie_List.dat')
fieldnames = ['Score_Rotten', 'Score_Audience', 'Rating', 'Genre', "Title"]

with open('Tomatoes.csv', 'w', encoding = 'UTF8', newline = '') as f:
    writer = csv.DictWriter(f, fieldnames = fieldnames)
    writer.writeheader()
    for movies in movie_list:
        try:
            movie_scraper = MovieScraper(movie_title=movies)
            movie_scraper.extract_metadata()
            scraped = movie_scraper.metadata
            scraped.update({"Title": movies})
            writer.writerow(scraped)            
        except:
            pass
```
